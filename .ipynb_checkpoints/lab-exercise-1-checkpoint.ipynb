### Exercise I

##### Python programs for simple arithmetics

```python
g= float(input("Input a number"))
y= 17
x= g-y
if y>g:
    print(abs(x))
else:
    print(x*2)
```

### Exercise II
##### Sum of three given nubers with conditionals

```python
x= float(input("Input a number"))
y= float(input("Input a number"))
z= 3
sum= x+y+z
if x!= 3 and y!= 3:
    print(sum)
else:
    print(sum*3)
```

### Exercise III
##### Truth value of given integers

```python
a= float(input("Input first number"))
b= float(input("Input second number"))
sum= a+b
sub= a-b
if sum== 5:
    print(True)
else:
    if sub== 5:
        print(True)
```

### Exercise IV
##### Sorting three Integers 

```python
x= float(input("Input first number"))
y= float(input("Input second number"))
z= float(input("Input third number"))
a= max(x,y,z)
b= min(x,y,z)
c= (x+y+z)-a-b
print(b)
print(c)
print(a)
```
